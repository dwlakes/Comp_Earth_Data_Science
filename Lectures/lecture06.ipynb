```python
import numpy as np
```

```python
thetas_in_degrees = range(0,360,5)
thetas_in_radians = np.radians(thetas_in_degrees)
sines = np.sin(thetas_in_radians)
sines
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(thetas_in_degrees, sines)
```

```python
plt.plot(thetas_in_degrees, sines, 'g-', label='Sine')
plt.xlabel('Degrees')
plt.ylabel('Sine')
plt.xlim(0, 360)
plt.title('Sine Curve')
plt.legend()
```

```python
cosines = np.cos(thetas_in_radians)
plt.plot(thetas_in_degrees, sines, 'r', label='Sine')
plt.plot(thetas_in_degrees, cosines, 'b--', label='Cosine')
plt.xlabel('Degrees')
plt.ylabel('Trig functions')
plt.xlim([0, 360])
plt.legend(loc=3)
```
