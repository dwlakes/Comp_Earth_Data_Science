```python
%matplotlib inline
```

```python
thetas_in_degrees = range(0,360,5)
thetas_in_radians = np.radians(thetas_in_degrees)

tans = np.tan(thetas_in_radians)
tans
```

```python
plt.plot(thetas_in_radians, tans, color='g', linestyle = 'dashdot', linewidth = 5, label='Tangent', )
plt.xlabel('Angles (radians)')
plt.ylabel('Tangents')
plt.ylim(-1,1)
plt.title('Plot of Tangents')
plt.legend()
plt.show()
```
